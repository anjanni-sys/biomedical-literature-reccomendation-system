# Biomedical Literature Recommendation System

This project builds a content-based recommendation system using NLP techniques to find similar biomedical research papers based on textual similarity.

## Problem Statement

With the rapid growth of biomedical research papers, it is difficult to find relevant literature efficiently. This project aims to recommend similar research papers based on their textual content using machine learning techniques.

## Dataset

- Source: PubMed abstracts dataset  
- Number of documents: ~13,000  
- Contains text from different biomedical domains such as deep learning, COVID-19, neuroprosthetics, etc.

## Workflow

1. Data preprocessing (handling missing values, combining text columns)
2. Text cleaning (removing symbols, lowercasing)
3. Feature extraction using TF-IDF
4. Similarity computation using cosine similarity
5. Recommendation system implementation

## Step 1: Data Preprocessing
We combine multiple text columns into a single column for analysis.

```python
import pandas as pd
import numpy as np
df=pd.read_csv("pubmed_abstracts.csv")
df.info()
df.describe()
df.shape
df_text = df.iloc[:, 1:9]  # only text columns (ignore links)
df_text = df_text.fillna('')

df['combined_text'] = df_text.apply(lambda x: ' '.join(x), axis=1)

df[['combined_text']].head()
df_text.head()
df.head()
```

## Step 2: Text Cleaning
We clean the text by removing special characters and normalizing it.

```python
import re

def clean_text(text):
    text = str(text)  # ensure string
    text = text.lower()
    text = text.replace('[', '').replace(']', '').replace("'", '')
    text = re.sub(r'[^a-zA-Z ]', ' ', text)   # replace with SPACE, not empty
    text = re.sub(r'\s+', ' ', text).strip()  # fix multiple spaces
    return text

df['clean_text']=df['combined_text'].apply(clean_text)

df['clean_text'].head()
```

## Step 3: Feature Extraction
TF-IDF is used to convert text into numerical vectors.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer= TfidfVectorizer(max_features=2000 , stop_words='english')
X= vectorizer.fit_transform(df['clean_text'])
X.shape
```

## Step 4: Similarity Calculation
Cosine similarity is used to measure similarity between documents.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(X)
similarity.shape
```

## Step 5: Recommendation System
We retrieve top similar documents based on similarity scores.

```python
def recommend(index):
    scores=list(enumerate(similarity[index]))
    scores=sorted(scores , key= lambda x: x[1] ,reverse=True)

    print("Top similar documents :\n")
    for i in scores[1:6]:
        print(f"Similarity Score: {i[1]:.3f}")
        text = df.iloc[i[0]]['combined_text']
        text = text.replace('[', '').replace(']', '').replace('(', '').replace(')', '')
        
        print(text[:200])
        print("------")
```

## Results

The system successfully recommends similar biomedical research papers based on textual similarity using TF-IDF and cosine similarity.

## Conclusion

This project demonstrates how NLP techniques can be applied to biomedical data to build a recommendation system. The model effectively identifies similar research papers and can assist researchers in literature discovery.

```python
recommend(10)
```
